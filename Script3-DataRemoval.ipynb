# Removal of unnecessary variables

Missing records:
Our target variable is PosIntFinal and we can't directly impute it without knowledge and hence removing those records from further analysis.
Removing records that can't be imputed without any background knowledge.

```python
TBI = TBI.drop(columns=['AgeInMonth','AgeTwoPlus', 'EmplType', 'Certification', 'High_impact_InjSev', 'Dizzy',
                        'EDCT', 'PosCT', 'EDDisposition', 'Observed', 'HospHead', 'DeathTBI', 'Neurosurgery', 
                        'Intub24Head', 'HospHeadPosCT'])
```

```python
TBI = TBI[~TBI['PosIntFinal'].isna()]
TBI = TBI[~TBI['Gender'].isna()]
```

```python
TBI.loc[TBI['Ethnicity'].isna(),'Ethnicity'] = 'Non-Hispanic'
TBI.loc[TBI['Race'].isna() & TBI['Ethnicity'].eq('Hispanic'),'Race'] = 'White'
TBI.loc[TBI['ActNorm'].isna(), 'ActNorm'] = 'Yes'
```

```python
TBI = TBI.drop(columns=[col for col in list(TBI.columns) if col.startswith('Ind')])
```

```python
print(f" % of NA's in each colum is \n\n{TBI.isnull().sum()/TBI.shape[0] * 100} appropriate action required")
```

```python
TBI = TBI.reindex(sorted(TBI.columns), axis=1)
```
