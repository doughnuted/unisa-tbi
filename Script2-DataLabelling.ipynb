We know that almost every field is a categorical variable and pandas is wrongly identified as float64 and hence we are transforming it to categrorical type except for age.

```python
for col in list(TBI):
        TBI[col] = TBI[col].astype(float).astype('Int64')
        if col not in ['AgeinYears', 'AgeInMonth']:
            TBI[col] = TBI[col].astype('category')
```
